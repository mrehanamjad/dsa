### Reverce Array


**Method I:**  
T-C = O($N^2$)  
S-C = O(N)

```python
def reverseArray(nums):
    if len(nums) <= 1:
        return nums
    return [nums.pop()] + reverseArray(nums)

print(reverseArray([5,7,3,2,6,1,5,9]))
```

**Method II:**  
T-C = O(N/2) = O(N)  
S-C = O(N)

```python
def reverseArrayWithSwap(nums,i):
    if nums[len(nums)//2] == i:
        return nums
    
    nums[i-1],nums[-i] = nums[-i],nums[i-1]

    return reverseArrayWithSwap(nums,i+1)

print(reverseArrayWithSwap([5,7,3,2,6,1,5,9],1))
```
