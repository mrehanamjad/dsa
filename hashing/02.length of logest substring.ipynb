```python
def lssLength(s):
    hash_list = []
    sub_str = ""
    for c in s:
        if c in sub_str:
            hash_list.append(len(sub_str))
            sub_str = c
        else:
            sub_str += c
    
    hash_list.append(len(sub_str))


    longest = 0
    for num in hash_list:
        if num > longest:
            longest = num
        
    return longest
lssLength("dvdf")
    
```
