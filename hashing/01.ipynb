### Q. 
`n = [5,3,22,1,5,5,7,5,10]`  
`m = [10,111,1,9,5,67,2]`   
**Constrains**
- 1 <= `n[1]` <= 10
- `n` can have 10^8 elements
- `m` can have 10^8 elements

**Sol 1 (Without hashing)**  
Time C = `O(n*m)`  
as `n` & `m` can have 10^8 elements so in worst case  
`O(n*m) = O(10^8 * 10^8) = O(10^16) => TLE Error`  
Space C = O(1)

```python
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

for num in m:
    count = 0
    for x in n:
        if x==num:
            count += 1
    print(f"{num} is {count} times in n ")
```

**Sol 1 (With hashing)**  
Time C = `O(n+m)`  
as `n` & `m` can have 10^8 elements so in worst case  
`O(n+m) = O(10^8 + 10^8) almost= O(10^8) => No TLE Error`  
Space C = `O(11)(constant so) => O(1)`

```python
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]
hash_list = [0]*11 # as gven condition =  n values can not be less than 1 and greater than 10
# we can also use dictionary here 
# if condition not given and n can have any value then the only option is dictionary 
for num in n: # O(N)
    hash_list[num] += 1
for num in m: # O(M)
    if num < 1 or num > 10:
        print(f"{num} is {0} times in n")
    else:
        print(f"{num} is {hash_list[num]} times in n")
    
```

```python
def twoSum(nums,target):
    hash_map = {}
    for i,num in enumerate(nums):
        remaining = target - num
        if remaining in hash_map:
            return [hash_map[remaining],i]
        hash_map[num] = i
    
    return []

print(twoSum([2,7,11,15],9))
```
