# 121. Best Time to Buy and Sell Stock
- `Easy`
- [Link](https://leetcode.com/problems/best-time-to-buy-and-sell-stock/)


```python
def MaxProfit(a):
    n = len(a)
    lower = a[0]
    maxProfit = 0

    for i in range(n):
        if a[i] < lower:
            lower = a[i]
            pass

        profit = a[i] - lower
        
        if maxProfit < profit:
            maxProfit = profit
        
    return maxProfit
```

```python
# TestCases
print(f"Test Case [7,1,5,3,6,4]: {MaxProfit([7,1,5,3,6,4])}")  # Output: 5
print(f"Test Case [7,6,4,3,1]: {MaxProfit([7,6,4,3,1])}")    # Output: 0
print(f"Test Case [1,2,3,4,5]: {MaxProfit([1,2,3,4,5])}")    # Output: 4
print(f"Test Case [10,5,20,15,30]: {MaxProfit([10,5,20,15,30])}") # Output: 25
print(f"Test Case [2,4,1]: {MaxProfit([2,4,1])}")          # Output: 2
```
